```python
import pandas as pd
import os
from PIL import Image
import requests
from io import BytesIO
import os
```

```python
if os.path.exists('data/prcsd_labels_df'):
    df = pd.read_pickle('data/prcsd_labels_df')
else:
    print('Please run the notebook for processing the labels first!')
```

## Downloading and scaling the images
In the current version here, each image is resized to 299x299 pixels. I've omitted colour depth checks since the images all appear to have 8bits per colour (which is fine).

# THE DOWNLOAD WILL TAKE >3 HOURS!!!

```python
if not os.path.exists('data/images'):
    os.makedirs('data/images')
```

```python
counter = 0
for i, url in enumerate(df.img_url.values):
    response = requests.get(url)
    im = Image.open(BytesIO(response.content))
#     The commented lines below would be needed if ratios are to be kept (but then cropping would be needed)
#     h = im.height
#     w = im.width
#     ratio = 299/(min(h,w))
    im = im.resize((299, 299))
    im.save('data/images/' + str(i) + '.png')
    counter += 1
    print(str(counter) + ' Images processed')
```

Each image was stored with the index of the row containing its URL in the df as name (1.png ... 14033.png)
